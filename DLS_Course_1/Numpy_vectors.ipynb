```python
import numpy as np

a = np.random.randn(5)
```

```python
print(a)
```

```python
print(a.shape) # rank 1 array => 이러한 데이터 구조 비추천
```

```python
print(a.T) # a == a.T
```

```python
print(np.dot(a, a.T))
```

```python
print(a.T)
```

```python
print(np.dot(a, a.T))
```
